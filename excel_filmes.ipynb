# Decisão sobre dados incorretos/ Faltantes/ Divergentes

```python
df.shape
```

```python
# Criar um novo DataFrame sem os Relançamentos
df_novo = df[df['Data de Lançamento'] != 'relançamento'].copy()
```

```python
df_novo.head()
```

```python
df_novo['Data de Lançamento'].value_counts()
```

```python
df_novo1 = df_novo[df_novo['Data de Lançamento'] != 'relançamento'].copy()
```

```python
df_novo1.dtypes
```

```python
df_novo1['Data de Lançamento'] = pd.to_datetime(df_novo1['Data de Lançamento'])
```

```python
df_novo1.dtypes
```

```python
df_novo1['Data de Lançamento'].shape
```

```python
# Quais os anos tivemos mais filmes lançados
df_novo1['Data de Lançamento'].dt.year.value_counts()
```

```python
# Plotar um gráfico com os filmes exibidos por ano
df_novo1['Data de Lançamento'].dt.year.value_counts().plot.bar(title = 'Exibição por Ano')
```

```python
# Qual o filme que teve maior bilheteria
df_novo1[df_novo1['Renda (R$) na semana dos dados'] == df_novo1['Renda (R$) na semana dos dados'].max()]
```

```python
#Os três filmes com maior bilheteria
df_novo1.nlargest(3, 'Renda (R$) na semana dos dados')
```

```python
# Qual o filme com menor bilheteria
df_novo1[df_novo1['Renda (R$) na semana dos dados'] == df_novo1['Renda (R$) na semana dos dados'].min()]
```
